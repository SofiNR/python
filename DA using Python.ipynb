```python
#Personal Attempt Before Doing The Assignment
import pandas as pd

df = pd.read_csv('Documents/DSandAI/Assignments/3/Titanic_train.csv')

print(df.info(),'\n\n')

print(df.describe(),'\n\n')

print(df.head(),'\n\n')

print(df.tail(),'\n\n')

print('Count of Females = ',df[df['Sex']=='female'].shape[0])

print('Count of Males = ',df[df['Sex']=='male'].shape[0])

df_group = df.groupby(['Sex','Survived'],sort=False)

# for name,group in df_group:
#     print(name,'\n',group,'\n\n')
    
print('\n\nSurvived Females = ',df_group.get_group(('female',1)).shape[0])

print('Survived Males = ',df_group.get_group(('male',1)).shape[0])

female_survival_rate = (df_group.get_group(('female',1)).shape[0]/df[df['Sex']=='female'].shape[0])*100

male_survival_rate = (df_group.get_group(('male',1)).shape[0]/df[df['Sex']=='male'].shape[0])*100

print('\nFemale Survival Rate = ',female_survival_rate)

print('\nMale Survival Rate = ',male_survival_rate)

print('\nTotal Survived = ',df['Survived'].sum())

print('\nAggregate Operations on DataFrame[\'Survived\'] \n',df['Survived'].agg(['min','mean','max','count','sum','median','std','var']))

print('\nOut of total ',df['Survived'].count(),' passengers, only ',df['Survived'].sum(),' survived')

survival_rate = (df['Survived'].sum()/df['Survived'].count())*100

print('\nSurvival Rate is ',survival_rate)

df_sorted = df.sort_values(['Sex','Survived'], axis=0, ascending=[True,False])

print('\n\nHead\n\n',df_sorted.head(10),'\n\nTail\n\n',df_sorted.tail(10))
```

```python
#Assignment 3 - Python for Data Analysis
# Q5 : Aggregation
# Q6 : GroupBy
# Q7 : OrderBy
# Q8 : Indexing
import pandas as pd

df = pd.read_csv('Documents/DSandAI/Assignments/3/Titanic_train.csv')

print('\n--------------------Q5 : Aggregation--------------------')
print('\n\nAverage Fare paid by the Passengers:')
print('\ndf.Fare.mean() = ',df.Fare.mean())

# Below works as well

# print('\ndf[\'Fare\'].mean() = ',df['Fare'].mean())
# print('\ndf.Fare.agg(\'mean\')',df.Fare.agg('mean'))
# print('\ndf[\'Fare\'].agg(\'mean\')',df['Fare'].agg('mean'))

# Personal Attempt
# print('\n\nAverage Fare paid by passengers in each Class of Travel :\n')
# print(df.groupby('Pclass',sort=False)[['Fare']].mean())

print('\n--------------------Q6 : GroupBy--------------------')
# From the info() of DataFrame we understand that Age columns have missing values
print('\n\nAverage Age of Passengers in each Class of Travel:\n')

df['Age'].fillna(0,inplace=True)

print(df.groupby('Pclass',sort=False)[['Age']].mean())

print('\n--------------------Q7 : OrderBy--------------------')
print('\n\nDataFrame Sorted by Fare Descending:\n')

df.sort_values('Fare',ascending=False,inplace=True)

print(df.head())

print('\n--------------------Q8 : Indexing--------------------')
print('\n\nName and Ticket Number of SUrvived Passengers:\n')

print(df.loc[(df.Survived == 1)][['Name','Ticket']])
```

```python
#Assignment 3 - Python for Data Analysis
# Q9  : Slicing
# Q10 : GroupBy
import pandas as pd

df = pd.read_csv('Documents/DSandAI/Assignments/3/Titanic_train.csv')

print('\n--------------------Q9 : Slicing--------------------')
print('\n\nDetails of Passengers Embarked from C and paid Fare > $50 :\n')

df['Embarked'].fillna('No Details',inplace=True)

df_subset = df.loc[(df.Embarked == 'C') & (df.Fare > 50)]

print(df_subset.head())

print('\n--------------------Q10 : GroupBy--------------------')
print('\n\nAverage Fare paid by Male and Female Passengers from each of the Embarked Cities:\n')
print(df.groupby(['Embarked','Sex'],sort=True)[['Fare']].mean())
```
