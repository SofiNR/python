```python
import numpy as np
import pandas as pd

df = pd.read_csv('Documents/DSandAI/Assignments/Titanic-train.csv')

#Converts Series to numpy array after filling Null values as 0
fare = df['Fare'].fillna(0).to_numpy()

#Numpy Array Element Sum
TotalFare = np.sum(fare)

print('\nWhat is the total fare paid by all passengers? ',np.round(TotalFare,2))

print('\n---------------------------------------------------------------------')

#Converts Series to numpy array after filling Null values as 0
age = df['Age'].fillna(0).to_numpy()

#Using Series.min() and Series.max(), also getting the index using Series.argmin()  and Series.argmax()
print('\nYoungest Passenger Age: ',df['Age'].min(),' Its Index Value : ',df['Age'].argmin())

print('\nOldest Passenger Age: ',df['Age'].max(),' Its Index Value : ',df['Age'].argmax())

print('\nLets see who are those passengers:')

#Extracting the Rows from Dataframe where the Youngest and Oldest value matches the rows Age column
print('\nYoungest Passengers:\n\n',df[(df['Age']==df['Age'].min())])

print('\nOldest Passengers:\n\n',df[(df['Age']==df['Age'].max())])

#print(df.iloc[[df['Age'].argmin(),df['Age'].argmax()],:])

#print(df.loc[[df['Age'].argmin(),df['Age'].argmax()]])

print('\n---------------------------------------------------------------------')

print('\nMin Fare: ',df['Fare'].min(skipna=True),' Its Index Value : ',df['Fare'].argmin())

print('\nMax Fare: ',df['Fare'].max(),' Its Index Value : ',df['Fare'].argmax())

print('\nLets see who are those passengers:\n')

print('\nPassengers who paid Minimum Fare:\n\n',df[(df['Fare']==df['Fare'].min(skipna=True))])

print('\nPassengers who paid Maximum Fare:\n\n',df[(df['Fare']==df['Fare'].max())])

#print(df.iloc[[df['Fare'].argmin(),df['Fare'].argmax()],:])

#print(df.loc[[df['Fare'].argmin(),df['Fare'].argmax()]])

print('\n---------------------------------------------------------------------')

#Numpy subtract array
fareSubAge = np.subtract(fare,age)
                        
print('\nMin and Max values of the difference of Age from Fare Columns:\n')
    
print(fareSubAge.min(),' ',fareSubAge.max(),' ')

#print('\nIndex of Min Value is: ',np.argmin(fareSubAge),'\nIndex of Max Value is: ',np.argmax(fareSubAge))

#print(df.iloc[[258,851],[1,2,3,4,5,6,7,9,11]])

minFareSubAge = fareSubAge.min()

maxfareSubAge = fareSubAge.max()

IndexOfMins = np.where(fareSubAge==minFareSubAge)[0]

IndexOfMaxs = np.where(fareSubAge==maxfareSubAge)[0]

print('\nPassengers whose Fare-Age difference is the Min Value:\n\n',df.iloc[IndexOfMins,[1,2,3,4,5,6,7,9,11]])

print('\nPassengers whose Fare-Age difference is the Max Value:\n\n',df.iloc[IndexOfMaxs,[1,2,3,4,5,6,7,9,11]])

print('\n---------------------------------------------------------------------')

print('\n\nMultiply the \'SibSp\' column (number of siblings/spouses aboard) with the \'Parch\' column (number of parents/children aboard). What does the resulting column represent?\n')

SiblingOrSpouse = df['SibSp'].fillna(0).to_numpy()

ParentOrChldrn = df['Parch'].fillna(0).to_numpy()

result = np.multiply(SiblingOrSpouse,ParentOrChldrn)

#resultarray = np.array([SiblingOrSpouse,ParentOrChldrn,result])

#Tresultarray = resultarray.T

#print(Tresultarray[50:100,:])

print(np.count_nonzero(result),' Passengers Travelled with more members from their families, either with their \n(Siblings and Parents) OR \n(Siblings and Children) OR \n(Spouse and Children) OR \n(Spouse and Parents).')

#Transpose of the array is taken and then the index of nonzero elements are found, returned as a tuple
nonzeroIndexs = np.nonzero(result.T)

#tuple to 2darray and 2d flattened to 1d array to be used as row index for DataFrame
arr=np.asarray(nonzeroIndexs).flatten()

moreFamily = df.iloc[arr,[1,2,4,5,6,7,9,11]]

print('\nDetails of the passengers who travelled with more family members are:\n',moreFamily.to_string())

#print(moreFamily.shape)

print('\n---------------------------------------------------------------------')

#fare/pclass, division by zero not possible hence not filling null values with zero
pclass = df['Pclass'].to_numpy()

#check how many NaN or null present
#np.count_nonzero(np.isnan(pclass))
if np.count_nonzero(np.isnan(pclass))==0:
    print('No blank values in the Pclass array')
else:    
    print('Pclass array has blank values and the count is = ',np.count_nonzero(np.isnan(pclass)))

#print(fare.shape,'',pclass.shape)    
fareByPclass = np.divide(fare,pclass)
print('\n\n',fareByPclass)

```

```python
import numpy as np
import pandas as pd

df = pd.read_csv('Documents/DSandAI/Assignments/Titanic-train.csv')

#Converts Series to numpy array after filling Null values as 0
fare = df['Fare'].fillna(0).to_numpy()

#Converts Series to numpy array after filling Null values as 0
age = df['Age'].fillna(0).to_numpy()

#AgeFare = np.array([age,fare]).T
AgeFare = np.array([age,fare])

#print(fare.shape,'   ',age.shape,'   ',AgeFare.shape)
#print('\n',age[0:5],'\n\n',fare[0:5],'\n\n',AgeFare[:,0:5])

print('\nSize of the Age and Fare Array is: ',AgeFare.shape,' and its dimension is: ',AgeFare.ndim) 
print('\n----------------------------------------------------------------------------------------')

AgeFareM10 = np.multiply(AgeFare,10)

print('\nElement-wise multiplication on the 2D Age and Fare Array will push one decimal place of the elements in the array.\n\n',AgeFareM10[:,0:5])
#print('\n',age[0:5],'\n\n',fare[0:5],'\n\n',AgeFare[:,0:5],'\n\n',AgeFareM10[:,0:5])
print('\n----------------------------------------------------------------------------------------')

AgeFareT = AgeFare.T

print('\nSize of the Transposed Age and Fare Array is: ',AgeFareT.shape,' and its dimension is: ',AgeFareT.ndim) 
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv('Documents/DSandAI/Assignments/Titanic-train.csv')

#prints column names if all the elements of that column is null
#empty_columns = df.isnull().all(axis=0)
#print(empty_columns)

columnNamesList = []
    
for column_name in df:
    if(df[pd.isnull(df[column_name])].size > 0):
        columnNamesList.append(column_name)

print('\nColumns having Null values are: ',columnNamesList)    

#columnNameArr = np.array(columnNamesList)

#print(df.describe(include='all'))
dfDescribe = df.describe(include='all')

#print(type(dfDescribe.loc['top','Cabin']))
df['Age'].fillna(0,inplace=True)
df['Cabin'].fillna(dfDescribe.loc['top','Cabin'],inplace=True)
df['Embarked'].fillna(dfDescribe.loc['top','Embarked'],inplace=True)

columnNamesList = []
for column_name in df:
    if(df[pd.isnull(df[column_name])].size > 0):
        columnNamesList.append(column_name)

print('\nColumns having Null values after removing them are: ',columnNamesList)           
```
